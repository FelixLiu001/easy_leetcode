```python
# 题号：0054	螺旋矩阵
class Solution(object):
    def spiralOrder(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: List[int]
        """
        u, d, l, r = 0, len(matrix)-1, 0, len(matrix[0])-1
        ret = []
        while True:
            for j in range(l,r+1): ret.append(matrix[u][j])
            u += 1
            if u > d: break
            for i in range(u,d+1): ret.append(matrix[i][r])
            r -= 1
            if r < l: break
            for j in range(r,l-1,-1): ret.append(matrix[d][j])
            d -= 1
            if u > d: break
            for i in range(d,u-1,-1):
                ret.append(matrix[i][l])
            l += 1
            if r < l: break
        return ret
    
matrix = [[1,2,3],[4,5,6],[7,8,9]]
Solution().spiralOrder(matrix)
```

```python
# 题号：0048	旋转图像
class Solution(object):
    def rotate(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        n = len(matrix)
        assert(n == len(matrix[0]))
        for i in range(n // 2):
            for j in range((n+1) // 2):
                t = matrix[i][j]
                matrix[i][j] = matrix[n-1-j][i]
                matrix[n-j-1][i] = matrix[n-1-i][n-j-1]
                matrix[n-1-i][n-j-1] = matrix[j][n-1-i]
                matrix[j][n-1-i] = t
                
matrix = [[1,2,3],[4,5,6],[7,8,9]]
Solution().rotate(matrix)
matrix
```

```python
# 题目：0704 二分查找
class Solution(object):
    def search(self, nums, target):
        """nums: List[int], target: int, ret: int"""
        l, r = 0, len(nums)-1
        while l <= r:
            m = l + (r - l) // 2
            if nums[m] == target: return m
            elif nums[m] < target: l = m + 1
            else: r = m - 1
        return -1
    
assert(Solution().search([-1,0,3,5,9,12], 9) == 4)
```

```python
# 题目：0034 在排序数组中查找元素的第一个和最后一个位置
class Solution(object):
    def searchRange(self, nums, target):
        """nums: List[int], target: int, ret: (int, int)"""
        def _left_most():
            l, r = 0, len(nums)-1
            ret = -1
            while l <= r:
                m = l + (r - l) / 2;
                if nums[m] < target: l = m+1
                elif nums[m] == target:
                    ret, r = m, m-1      # left_most 和 right_most 只在这里不一样
                else: r = m-1
            return ret
        
        def _right_most():
            l, r = 0, len(nums)-1
            ret = -1
            while l <= r:
                m = l + (r - l) / 2;
                if nums[m] < target: l = m+1
                elif nums[m] == target:
                    ret, l = m, m+1      # left_most 和 right_most 只在这里不一样
                else: r = m-1
            return ret
        
        return _left_most(), _right_most()
```

```python
# 题目：0240 搜索二维矩阵 II
class Solution(object):
    def searchMatrix(self, matrix, target):
        """matrix: List[List[int]], target: int, ret: bool"""
        if not matrix: return False
        m, n = len(matrix), len(matrix[0])
        x, y = 0, n-1
        while x < m and y >= 0:
            if matrix[x][y] == target: return True
            elif matrix[x][y] > target: y -= 1
            else: x += 1
        return False
```

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
# 题号 0083 删除排序链表中的重复元素
class Solution(object):
    def deleteDuplicates(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head: return head
        ret, last, p = head, head, head.next
        while p:
            if p.val != last.val:
                last, p = p, p.next
            else:
                last.next, p = p.next, p.next
        return ret
```

```python
# 题号 0082 删除排序链表中的重复元素 II
class Solution(object):
    def deleteDuplicates(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        dummy = ListNode(0, head)
        p, n = dummy, dummy.next
        while n:
            if n.next and n.val == n.next.val:  # 需删除
                t = n.next
                while t and t.val == n.val: t = t.next
                n = t
                p.next = n
                continue
            else:
                p.next = n
                p, n = p.next, n.next
        return dummy.next
```

```python
# 题号 0206 反转链表
class Solution(object):
    def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        p = None
        while head:
            t = head.next
            head.next = p
            p = head
            head = t
        return p
```

```python
# 题号 0092 反转链表 II
class Solution(object):
    def reverseBetween(self, head, left, right):
        """
        :type head: ListNode
        :type left: int
        :type right: int
        :rtype: ListNode
        """
        def reverseUntil(start, end):
            pre = None
            while start and pre != end:
                t = start.next
                start.next = pre
                pre = start
                start = t
            return pre

        def forward(node, n):
            while n > 0 and node:
                node = node.next;
                n -= 1
            return node
                
        dummy = ListNode(0, head)
        pre = forward(dummy, left-1)
        if not pre: return dummy.next  # 边界检查
        start = pre.next
        end = forward(dummy, right)
        post = end.next if end else None
        reverseUntil(start, end)
        # 之后的 end 结点是开始结点，start 结点是结束结点
        pre.next = end
        start.next = post
        return dummy.next
```

```python
# 题号 0234	回文链表
class Solution(object):
    def isPalindrome(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        dat = []
        while head:
            dat.append(head.val)
            head = head.next
        return dat == dat[::-1]
```

```python
# 题号： 0148	排序链表
class Solution(object):
    def sortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head or not head.next: return head

        def _merge(h1, h2):
            dummy = ListNode(0, None)
            p = dummy
            while h1 and h2:
                if h1.val <= h2.val:
                    p.next = h1
                    p, h1 = p.next, h1.next
                else:
                    p.next = h2
                    p, h2 = p.next, h2.next
            if h1: p.next = h1
            if h2: p.next = h2
            return dummy.next
        
        def _find_mid(h):
            slow, fast = h, h
            while fast and fast.next and fast.next.next:
                slow, fast = slow.next, fast.next.next
            return slow
        
        mid = _find_mid(head)
        post = mid.next
        mid.next = None
        return _merge(self.sortList(head), self.sortList(post))
```

```python
# 题号：0021	合并两个有序链表
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        dummy = ListNode(0)
        p = dummy
        while list1 and list2:
            if list1.val <= list2.val:
                p.next = list1
                p, list1 = p.next, list1.next
            else:
                p.next = list2
                p, list2 = p.next, list2.next
        if list1: p.next = list1
        if list2: p.next = list2
        return dummy.next
```

```python
# 题号：0023	合并 K 个升序链表
import heapq
class Solution(object):
    def mergeKLists(self, lists):
        """
        :type lists: List[ListNode]
        :rtype: ListNode
        """
        if not lists: return None
        # [(val, idx), ...]
        data = [(li.val, idx) for idx, li in enumerate(lists) if li]
        if not data: return None
        if len(data) == 1: return lists[data[0][1]]
        heapq.heapify(data)
        dummy = ListNode(0)
        pre = dummy
        while len(data) > 1:
            _, idx = heapq.heappop(data)
            pre.next = lists[idx]
            pre = pre.next
            if lists[idx].next:
                lists[idx] = lists[idx].next
                heapq.heappush(data, (lists[idx].val, idx))
        pre.next = lists[data[0][1]]
        return dummy.next
```

```python
# 题号：0141	环形链表
class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        if not head or not head.next: return False
        slow, fast = head, head
        while fast and fast.next:
            slow, fast = slow.next, fast.next.next
            if slow == fast: return True
        return False
```

```python
# 题号：0142	环形链表 II
class Solution(object):
    def detectCycle(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head: return None
        slow, fast = head, head
        while fast and fast.next:
            slow, fast = slow.next, fast.next.next
            if slow == fast:
                re = head
                while slow != re:
                    slow, re = slow.next, re.next
                return re
        return None
```

```python
# 题号：0160	相交链表
class Solution(object):
    def getIntersectionNode(self, headA, headB):
        """
        :type head1, head1: ListNode
        :rtype: ListNode
        """
        def _getLength(head):
            len = 0
            while head:
                head = head.next
                len += 1
            return len
        def _forward(head, n):
            while n > 0 and head:
                head = head.next
                n -= 1
            return head

        la, lb = _getLength(headA), _getLength(headB)
        if la >= lb: headA = _forward(headA, la - lb)
        else: headB = _forward(headB, lb - la)
        while headA and headB:
            if headA == headB: return headA
            headA, headB = headA.next, headB.next
        return None
```

```python
# 题号：0019	删除链表的倒数第 N 个结点
class Solution(object):
    def removeNthFromEnd(self, head, n):
        """
        :type head: ListNode
        :type n: int
        :rtype: ListNode
        """
        if not head or n <= 0: return head
        dummy = ListNode(0, head)
        
        tail = dummy
        while n >= 0 and tail: tail, n = tail.next, n-1
        if n >= 0: return head

        pre = dummy
        while tail: pre, tail = pre.next, tail.next
        pre.next = pre.next.next
        return dummy.next
```

```python
# 剑指 Offer 22	链表中倒数第k个节点
class Solution(object):
    def trainingPlan(self, head, cnt):
        """
        :type head: Optional[ListNode]
        :type cnt: int
        :rtype: Optional[ListNode]
        """
        p = head
        while cnt > 0 and p:
            p = p.next
            cnt -= 1
        if cnt > 0: return None
        while p:
            head, p = head.next, p.next
        return head
```

```python
# 题号：0143	重排链表
class Solution(object):
    def reorderList(self, head):
        """
        :type head: ListNode
        :rtype: None Do not return anything, modify head in-place instead.
        """
        def GetMiddle(h):
            slow, fast = h, h
            while fast and fast.next and fast.next.next:
                slow, fast = slow.next, fast.next.next
            return slow
        def Reverse(h):
            p = None
            while h:
                t = h.next
                h.next = p
                p = h
                h = t
            return p
        def InterMerge(h1, h2):
            dummy = ListNode(0)
            p = dummy
            while h1 and h2:
                t_h1, t_h2 = h1.next, h2.next
                p.next = h1
                p = p.next
                p.next = h2
                p = p.next
                h1, h2 = t_h1, t_h2
            if h1: p.next = h1
            return dummy.next
        
        mid = GetMiddle(head)
        
        # 断开两个链表
        l2 = Reverse(mid.next)
        mid.next = None
        return InterMerge(head, l2)
```

```python
# 题号：0155	最小栈
class MinStack(object):

    def __init__(self):
        self._vals = []
        self._mins = []

    def push(self, val):
        """
        :type val: int
        :rtype: None
        """
        self._mins.append(val if len(self._mins) == 0 else min(val, self._mins[-1]))
        self._vals.append(val)

    def pop(self):
        """
        :rtype: None
        """
        self._mins.pop()
        return self._vals.pop()

    def top(self):
        """
        :rtype: int
        """
        return self._vals[-1] if len(self._vals) > 0 else None
    
    def getMin(self):
        """
        :rtype: int
        """
        return self._mins[-1] if len(self._mins) > 0 else None

# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

```python
# 题号：0020	有效的括号
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        st = ""
        for i in s:
            if i in ('[{('):
                st += i
                continue
            if i == ']' and st.endswith('[') or \
               i == '}' and st.endswith('{') or \
               i == ')' and st.endswith('('):
                st = st[:-1]
                continue
            else: return False
        return len(st) == 0
```

```python
# 题号：0232	用栈实现队列
class MyQueue(object):

    def __init__(self):
        self.in_stack = []
        self.out_stack = []

    def push(self, x):
        """
        :type x: int
        :rtype: None
        """
        self.in_stack.append(x)

    def pop(self):
        """
        :rtype: int
        """
        if self.out_stack: return self.out_stack.pop()
        if self.in_stack:
            while self.in_stack: self.out_stack.append(self.in_stack.pop())
            return self.out_stack.pop()
        return None

    def peek(self):
        """
        :rtype: int
        """
        if self.out_stack: return self.out_stack[-1]
        if self.in_stack:
            while self.in_stack: self.out_stack.append(self.in_stack.pop())
            return self.out_stack[-1]
        return None

    def empty(self):
        """
        :rtype: bool
        """
        return len(self.in_stack) == 0 and len(self.out_stack) == 0 


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```

```python
# 题号：0023	合并 K 个升序链表
import heapq
class Solution(object):
    def mergeKLists(self, lists):
        """
        :type lists: List[ListNode]
        :rtype: ListNode
        """
        if not lists: return None
        # [(val, idx), ...]
        data = [(li.val, idx) for idx, li in enumerate(lists) if li]
        if not data: return None
        if len(data) == 1: return lists[data[0][1]]
        heapq.heapify(data)
        dummy = ListNode(0)
        pre = dummy
        while len(data) > 1:
            _, idx = heapq.heappop(data)
            pre.next = lists[idx]
            pre = pre.next
            if lists[idx].next:
                lists[idx] = lists[idx].next
                heapq.heappush(data, (lists[idx].val, idx))
        pre.next = lists[data[0][1]]
        return dummy.next
```

```python
# 题号： 0144	二叉树的前序遍历
class Solution(object):
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root: return []
        return [root.val] + self.preorderTraversal(root.left) \
                          + self.preorderTraversal(root.right)
```

```python
# 题号： 0094	二叉树的中序遍历
class Solution(object):
    def inorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root: return []
        return self.inorderTraversal(root.left) + [root.val] \
               + self.inorderTraversal(root.right)
```

```python
# 题号： 0102	二叉树的层序遍历
class Solution(object):
    def levelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root: return []
        rets = []
        st = [root]
        while len(st) > 0:
            ret, next_st = [], []
            for node in st:
                ret.append(node.val)
                if node.left: next_st.append(node.left)
                if node.right: next_st.append(node.right)
            st = next_st
            rets.append(ret)
        return rets
```

```python
# 题号： 0103	二叉树的锯齿形层序遍历
class Solution(object):
    def zigzagLevelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root: return []
        arr, rets = [root], []
        left_to_right = 1
        while len(arr) > 0:
            next_arr = []
            for node in arr:
                if node.left: next_arr.append(node.left)
                if node.right: next_arr.append(node.right)
            arr = arr if left_to_right else arr[::-1]
            ret = [node.val for node in arr]
            arr = next_arr
            rets.append(ret)
            left_to_right = (left_to_right + 1) % 2
        return rets
```

```python
# 题号： 0236	二叉树的最近公共祖先
class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        if not root or root == p or root == q: return root
        left = self.lowestCommonAncestor(root.left, p, q)
        right = self.lowestCommonAncestor(root.right, p, q)
        if not left and not right: return None
        elif left and not right: return left
        elif not left and right: return right
        return root
```

```python
# 题号： 0104	二叉树的最大深度
class Solution(object):
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if not root: return 0
        ld = self.maxDepth(root.left)
        rd = self.maxDepth(root.right)
        return max(ld, rd) + 1
```

```python
# 题号： 0112	路径总和
class Solution(object):
    def hasPathSum(self, root, targetSum):
        """
        :type root: TreeNode
        :type targetSum: int
        :rtype: bool
        """
        if not root: return False
        if root.val == targetSum and \
            not root.left and \
            not root.right: return True
        val = targetSum - root.val
        return self.hasPathSum(root.left, val) or \
                self.hasPathSum(root.right, val)
```

```python
# 题号： 0113	路径总和 II
class Solution(object):
    def pathSum(self, root, targetSum):
        """
        :type root: TreeNode
        :type targetSum: int
        :rtype: List[List[int]]
        """
        if not root: return []
        rets = []
        def _path_sum(node, target, paths):
            if not node: return
            if node.val == target and \
                not node.left and \
                not node.right:
                rets.append(paths + [node.val])
                return
            val = target - node.val
            _path_sum(node.left, val, paths + [node.val])
            _path_sum(node.right, val, paths + [node.val])

        _path_sum(root, targetSum, [])
        return rets
```

```python
# 题号： 0101	对称二叉树
class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if not root: return True

        def isDoubleSymmetric(left, right):
            if not left and not right: return True
            elif not left and right or \
                 left and not right: return False
            if left.val != right.val: return False
            return isDoubleSymmetric(left.left, right.right) and \
                    isDoubleSymmetric(left.right, right.left)

        return isDoubleSymmetric(root.left, root.right)
```

```python
# 题号： 0124	二叉树中的最大路径和
class Solution(object):
    def __init__(self):
        self.ret = None

    def _max_path_sum(self, root):
        if not root: return 0
        a, b, c = root.val, self._max_path_sum(root.left), \
                            self._max_path_sum(root.right)
        if self.ret == None: self.ret = a + b + c
        else: self.ret = max(self.ret, a + b + c)
        return max(0, a + b, a + c)

    def maxPathSum(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        self.ret = None
        self._max_path_sum(root)
        return self.ret
```

```python
# 题号： 0199	二叉树的右视图
class Solution(object):
    def rightSideView(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root: return []
        arr, rets = [root], []
        while len(arr) > 0:
            next_arr = []
            for node in arr:
                if node.right: next_arr.append(node.right)
                if node.left: next_arr.append(node.left)
            rets.append(arr[0].val)
            arr = next_arr
        return rets
```

```python
# 题号： 0226	翻转二叉树
class Solution(object):
    def invertTree(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        if not root: return root
        t_left, t_right = root.left, root.right
        root.left = self.invertTree(t_right)
        root.right = self.invertTree(t_left)
        return root
```

```python
# 题号：0024	两两交换链表中的节点
class Solution(object):
    def swapPairs(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        dummy = ListNode(0, head)
        pre = dummy
        while pre.next and pre.next.next:
            p1, p2 = pre.next, pre.next.next
            post = p2.next
            pre.next = p2
            p2.next = p1
            p1.next = post
            pre = p1
        return dummy.next
```
